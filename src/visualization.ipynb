```python
# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('../results/step7_results.csv') 
# Агрегируем
agg = df.groupby('model').agg(
    smape_mean=('smape','mean'),
    smape_std=('smape','std')
).reset_index()

# Строим barplot
plt.figure(figsize=(10,6))
sns.barplot(data=agg, x='model', y='smape_mean', yerr=agg['smape_std'])
plt.title('Model comparison: mean sMAPE with std')
plt.ylabel('sMAPE')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
